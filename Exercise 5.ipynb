```python
import numpy as np
import random as rnd
import scipy
```

```python

def prob_matrix(n,p):
    C = np.zeros((n,n))
    for row in range(n):
        for col in range(n):
            probability=rnd.uniform(0.0,1.0)
            C[row][col]=1 if probability <=p else 0
    return C
```
